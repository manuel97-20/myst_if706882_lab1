# Introducción

### Las estrategias de inversión están en un constante debate en cuanto a si se debe seguir una estragia activa, realizar rebalanceos constantes de acuerdo a medidas de desempeño, ratios etc... o una pasiva, es decir, indexarse al S&P 500 o al IPC y solo esperara a que la inversión madure (conforme al horizonte de inversión de cada persona en particular) bajo la premisa de que eventualmente el mercado se alzará. 

### Se ha encontrado que en tiempos difíciles para los índices, recesiones y depresiones económicas, el público inversionista busca ayuda de la mano del experto, que pueda sobre llevar su dinero por ese camino lleno de baches, alta volatilidad y expectativas negativas. La inversión pasiva aún así no ha perdido demasiado terreno, cada vez se vuelve una opción viable como inversión a largo plazo, y claro que es atractiva pues hay menos comisiones por manejo de fondo al gestor y muchas menos comisiones por transacción. "No creo que sea fácil decir que la inversión activa es bueno por esto o que la inversión pasiva es buena por esto otro, existen diferente ventajas para cada una" menciona Rick Boebel, CFA. Para dar un ejemplo de lo anterior: los mercados emergentes se han vuelto un atractivo para ambas estrategias, porque existe claro un mayor riesgo, pero al mismo tiempo una oportunidad de batir a los índices claásicos (S&P 500, FTSE 100, DOW JONES, etc...),  ahora se puede pensar que en economías en vías de desarrollo existan ineficiencias de información que los gestores de estrategia podrían aprovechar y obtener mejores rendimientos, sin embargo, tambíen es posible que en este tipo de mercados los activos sean mucho menos líquidos y que al hacer transacciones se incurra en costos para el inversionista, lo cual volvería atractiva la gestión pasiva.

 # Procesos 

### De los archivos descargados de BlackRock se obtienen las fechas en las que nos interesará conocer la posición tanto de la inversión pasiva como activa

```python
fechas = fn.f_fechas(p_archivos=dt.archivos)
print(fechas['t_fechas']) # usaremos las siguientes fechas para lolizar los precios en yahoo finance 
```

### Al revisar los archivos con detalle, se observa que el NAFTRAC incorpora y reemplaza las empresas que lo conforman mes a mes, sin embargo, el objetivo en esta ocasión es verificar lo que pasa con el capital sin cambiar de empresas, desde que se decide realizar la inversión, se tomarán todos los tickers únicos de cada mes del NAFTRAC y serán usados para descargar los precios. Aquí mismo se retira el Ticker de Coca-Cola Femsa pues para este periodo de tiempo en particular la acción sufrió un cambio y no se tomará en cuenta para el desarrollo de la inversión


```python
global_tickers = fn.f_global_tickers(dt.data_archivos, dt.filenames)
print(global_tickers)
```

### Una vez que se tienen los tickers globales, se pasa a descargar los precios desde yahoo finance 

```python
prices = fn.prices_download(global_tickers) # se descargan los precios diarios de los tickers globales 
prices.head()
```

### Al revisar el diccionario de archivos realizado para organizar los archivos descargados de BlackRock, se observó que los no se encontraban en orden cronólogico, la solución entonces fue hacer una lista con las fechas de los archivos y posteriormente reemplazar la llave de cada archivo por su 

```python
dates_list = [datetime.datetime.strptime(i, "%d%m%y").date() for i in fechas['i_fechas']]
for i in range(len(fechas['i_fechas'])):
    for j in range(32):
        if fechas['i_fechas'][i] in dt.filenames[j]:
            dt.data_archivos[fechas['i_fechas'][i]]=dt.data_archivos.pop(dt.filenames[j])
            dt.data_archivos[dates_list[i]]=dt.data_archivos.pop(fechas['i_fechas'][i])
```

```python
dt.data_archivos[dates_list[0]] #aquí ya se puede observar que el primer archivo coincide con la fecha en la que colocaremos la inversión
```

### Como se han descargado de manera diaria los precios, lo que interesa es conocer cuanto costaba la acción en cada una de las fechas de interés, por lo que se limpiará lo descargado de yahoo finance y tomaremos únicamente los precios de cierre 

```python
# preparar la tabla que usaremos para ir cotejando los precios de todos los meses
clean_prices, data_close = fn.clean_price(data_archivos=dt.data_archivos, dates_list=dates_list, data=prices, t_fechas=fechas['t_fechas'])
clean_prices.head()

# ya solo se cuenta con precios de cierre 
```

### inicializamos la inversión pasiva, con los tickers y pesos del archivo del 31-01-2018

```python
pos_datos = fn.pasiva_ini(data_archivos=dt.data_archivos, dates_list=dates_list, precios=clean_prices)
pos_datos.tail()
```

### obetenemos la tabla final con los resultados y la posición pasiva mes a mes, se presentan los últimos 5 meses donde se observa que acumulado, el rendimiento para este periodo ha sido de -21%

```python
results = fn.inv_pasiva(pos_datos, dates_list, clean_prices)
results.tail() 
```

### Se inicializa la inversión activa, donde en este caso haremos operaciones de acuerdo al siguiente mandato: tomando el activo con la ponderación del archivo de BlackRock del 31-01-2018, se destinará a ese activo el 50% de lo asignado en la pasiva. La acción con la poderación más alta fue Amércia móvil, siguiente a esto se realizará una compra del 10% del efectivo restante hasta que finalmente se agote, la señal para comprar vendrá a ser dada cada que vez que el precio de cierre este por debajo 1% del precio de apertura del mismo día y la compra se realizará al precio de apertura del día siguiente 

```python
pos_activa, cash_activa, cash_ini = fn.activa_ini(dt.data_archivos,dates_list,clean_prices)
pos_activa.head()
```

### Obtenemos el histórico de operaciones en la inversión activa, según el mandato previamnete descrito 

```python
historico, saved_cash =  fn.operaciones(activa_ini=pos_activa,data=prices,data_close=data_close,
                                        t_fechas=fechas['t_fechas'],cash_activa=cash_activa)
historico #se puede observar el que efectivo se agotó el 19-02-2018
```

### se obtiene la tabla mensual con la posisicón mes a mes de la estrategia activa, podemos ver que prácticamente tuvimos el mismo rendimiento acumulado que con la inversión pasiva 

```python
inv_activa = fn.inv_activa(historical=historico,activa_ini=pos_activa,dates_list=dates_list,precios=clean_prices
                           ,saved_cash=saved_cash,cash_ini=cash_ini)
inv_activa.tail()
```
